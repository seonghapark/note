```python
### pip3 install act-atmos mpl2nc arm_pyart

import act
import matplotlib.pyplot as plt

import numpy as np

print(np.__version__)

import glob
```

```python
def pplot(i):
    pobj = act.io.mpl.read_sigma_mplv5(i)

    display = act.plotting.TimeSeriesDisplay(
        {'MiniMPL': pobj},
        subplot_shape=(1,),
        figsize=(15, 5),
    )

    # Plot PSL Radar followed by the parsivel data.
    display.plot(
        'nrb_crosspol',
        dsname='MiniMPL',
        cmap='act_HomeyerRainbow',
        subplot_index=(0,),
    )

    plt.show()

files = glob.glob('minimpl_202203/*')
c = 0
for i in files:
    print(i, c)
    pplot(i)
    c += 1
    if c == 15:
        break
```
